# Top 10 highest-rated iPhone on Flipkart in India?

```python
Highest_rated = data.sort_values(by=['Star Rating'], ascending=False)
Highest_rated = Highest_rated.head(10)
```

```python
print(Highest_rated['Product Name'])
```

```python
print(Highest_rated)
```

# How many rating do the highest-rated iPhone on Flipkart have?

```python
iPhone = Highest_rated['Product Name'].value_counts()
print(iPhone)
```

```python
label = iPhone.index
counts = Highest_rated['Number Of Ratings']
figure = px.bar(Highest_rated, x=label, y = counts,title = "Number of Ratings Highest Rated iPhone")
figure.show()
```

# Which iPhone has the highest number of reviews on Flipkart?

```python
label = iPhone.index
counts = Highest_rated['Number Of Reviews']
figure = px.bar(Highest_rated, x=label, y = counts,title = "Number Of Reviews Highest Rated iPhone")
figure.show()
```

# What is the relationships between the sales price of iPhone and the number of rating on Flipkart?

```python
data.head(2)
```

```python
figure = px.scatter(data_frame =data, x ="Number Of Ratings",y = "Sale Price",size='Discount Percentage',trendline='ols',title='Relationship Between Sale Price and Number of Ratings')
figure.show()
```

There is a negative linear relationship between the sales price of iPhone and the number of ratings. It means iPhones with lower sales prices are sold more in india.

# What is the relationships between the Discount percentage of iPhone and the number of rating on Flipkart?

```python
figure = px.scatter(data_frame = data, x = "Number Of Ratings",y = "Discount Percentage", size = "Sale Price", trendline = "ols", title = "Relationship Between Discount Percentage and Number of Ratings")
figure.show()
```

# Most Expensive and least expensive

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

#Display the results
print("Most Expensive iPhone:")
print(most_expensive)

print("\nLeast Expensive iPhone:")
print(least_expensive)
```
